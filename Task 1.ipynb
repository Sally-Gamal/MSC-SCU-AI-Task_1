```python
import time


def get_valid_number(message):
    while True:
        try:
            val = int(input(message))
            if 0 <= val <= 30:
                return val
            else:
                print("โ ูุฑุฌู ุฅุฏุฎุงู ุฑูู ุฏูุฑ ุตุญูุญ (ูู 0 ูู 30)")
        except ValueError:
            print("โ ุฅุฏุฎุงู ุฎุงุทุฆ! ุฃุฑูุงู ููุท.")

def full_elevator_system():
    current_floor = 0
    waiting_list = []

    
    print("--- ูุธุงู ุงููุตุนุฏ ุงูุฐูู ---")
    count = get_valid_number("ูู ุนุฏุฏ ุงูุฃุดุฎุงุต ุงูุฐูู ููุชุธุฑูู ุงููุตุนุฏุ ")

    for i in range(count):
        print(f"\nุจูุงูุงุช ุงูุดุฎุต ุฑูู {i+1}:")
        name = input("ุงูุงุณู: ")
        start = get_valid_number(f"ูู ุงูุฏูุฑ: ")
        dest = get_valid_number(f"ุฅูู ุงูุฏูุฑ: ")
        
       
        waiting_list.append({"name": name, "from": start, "to": dest})

    print("\nโ ุชู ุชุณุฌูู ุงูุทูุจุงุช. ุงููุตุนุฏ ุณูุจุฏุฃ ุงูุนูู ุงูุขู...")
    time.sleep(1)


    while len(waiting_list) > 0:
        
        print(f"\n ุงููุตุนุฏ ุญุงููุงู ูู ุงูุฏูุฑ: {current_floor}")

       
        best_index = 0
        min_dist = abs(waiting_list[0]["from"] - current_floor)
        
        for i in range(1, len(waiting_list)):
            dist = abs(waiting_list[i]["from"] - current_floor)
            if dist < min_dist:
                min_dist = dist
                best_index = i
        
    
        person = waiting_list.pop(best_index)
        
        p_name = person["name"]
        p_start = person["from"]
        p_dest = person["to"]

        
        if current_floor != p_start:
            print(f"๐ผ๐ฝ ุงููุตุนุฏ ูุชุญุฑู ููุฏูุฑ {p_start} ุนุดุงู ูุงุฎุฏ {p_name}...")
            time.sleep(1.5)
            current_floor = p_start
        
        print(f"๐ช {p_name} ุฑูุจ ุงููุตุนุฏ.")

       
        if current_floor != p_dest:
            print(f"๐ผ๐ฝ ุงููุตุนุฏ ุจููุตู {p_name} ููุฏูุฑ {p_dest}...")
            time.sleep(1.5)
            current_floor = p_dest
            
        print(f"๐ {p_name} ูุตู ููุฒู ุจุงูุณูุงูุฉ.")
        print("-" * 30)

    print("๐ ุงููุตุนุฏ ุฎูุต ูู ุงูุทูุจุงุช ููุงูู ูุณุชุฑูุญ.")


full_elevator_system()
```

```python
def smart_chatbot():
    print("๐ค Bot: I am ready! (Type 'bye' to exit)")
    
    rules = {
        "hello": "Hi! How can I help you?",
        "ai": "AI is the simulation of human intelligence."
        "name": "My name is Robo-Bot.",
        "bye": "Goodbye!"
    }

    while True:
        user_input = input("You: ").strip().lower()

        if user_input == "bye":
            print("Bot: Goodbye!")
            break

     
        found_match = False 
        
        for key in rules:
            
            if key in user_input:
                print(f"Bot: {rules[key]}")
                found_match = True
                break  
        
    
        if not found_match:
            print("Bot: Sorry, I don't understand.")


smart_chatbot()
```
