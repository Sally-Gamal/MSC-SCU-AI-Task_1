## Part 5: Array Operations

```python
arr10 = np.array([10, 20, 60, 40, 90])
print("Original Array:")
print(arr10)
grad = np.gradient(arr10)
print("\nGradient of arr10:")
print(grad)
```

```python
conditions = [
    arr10 < 10,                  
    (arr10 >= 10) & (arr10 <= 30), 
    arr10 > 30                    
]

choices = ['Low', 'Medium', 'High'] 

arr_classified = np.select(conditions, choices, default='Unknown')
print(arr_classified)
```

```python
arr_appended = np.append(arr10, [60, 70, 80])
print(arr_appended)
```

## Part 2: DataFrame Creation

```python
data = {
    'Name': ['Sally', 'Ali', 'Amany', 'Ahmad'],
    'Age': [18, 19, 20, 21],
    'Math_Score': [85, 72, 90, 65],
    'Gender': ['Female', 'Male', 'Female', 'Male']
}
```

```python
df = pd.DataFrame(data)
df
```

```python
df.info
```

```python
df.describe()
```

## Part 3: Indexing & Filtering

```python
df['Math_Score']
```

```python
df.loc[:, ['Name', 'Math_Score']]
```

```python
df.iloc[0:2, 1:3]
```

```python
high_scores = df[df['Math_Score'] > 70]
high_scores
```

## Part 4: Data Cleaning

```python
df.loc[3, 'Math_Score'] = np.nan
df
```

```python
mean_score = df['Math_Score'].mean()  
df['Math_Score'] = df['Math_Score'].fillna(mean_score)
df
```

```python
df = df.drop(columns=['Age'])
df
```

## Part 5: Apply & Transformation

```python
df['Result'] = df['Math_Score'].apply(lambda x: 'Fail' if x < 50 else 'Pass')
df
```

```python
gender_dummies = pd.get_dummies(df['Gender'], dtype=int)

df = pd.concat([df, gender_dummies], axis=1)
df = df.drop(columns=['Gender'])

df
```

## Part 6: Combining & Reshaping

```python
df2 = pd.DataFrame({
    'Name': ['Sara', 'Eman'],
    'Math_Score': [78, 88],
    'Result': ['Pass', 'Pass'],
    'Female': [1, 0],
    'Male': [0, 1]
})
```

```python
df2
```

```python
df_combined = pd.concat([df, df2], ignore_index=True)
df_combined
```

```python
df_combined.groupby('Female')['Math_Score'].mean()
```
